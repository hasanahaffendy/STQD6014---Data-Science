<h2> <strong> About Dataset: </strong> </h2>
This dataset contains a comprehensive list of the most popular songs in 2023, as listed on Spotify. The dataset contains many more features than similar datasets. It provides information about each song's characteristics, popularity, and presence across various music platforms. The dataset contains the track name, artist(s) name, release date, Spotify playlists and charts, streaming statistics, and Apple Music.


<h3> <strong> Key Features: </strong> </h3>

- track_name: Name of the song

- artist(s)_name: Name of the artist(s) of the song

- artist_count: Number of artists contributing to the song

- released_year: Year when the song was released

- released_month: Month when the song was released

- released_day: Day of the month when the song was released

- in_spotify_playlists: Number of Spotify playlists the song is included in

- in_spotify_charts: Presence and rank of the song on Spotify charts

- streams: Total number of streams on Spotify

- in_apple_playlists: Number of Apple Music playlists the song is included in

- in_apple_charts: Presence and rank of the song on Apple Music charts

Link: https://www.kaggle.com/datasets/nelgiriyewithana/top-spotify-songs-2023

```python
# Example URL
spotify = 'spotify-2023.csv'

# Read data directly from the URL
df = pd.read_csv(spotify, encoding='latin1')
```

```python
# Perform necessary cleaning steps
# Drop columns with NaN values
df_cleaned = df.dropna(axis=1)
df_cleaned
```

<h1><strong>2. Frequency Analysis:</strong></h1>

```python
# Frequency of songs by artist(s)
artist_frequency = df_cleaned['artist(s)_name'].value_counts()

# Frequency of songs by released year
year_frequency = df_cleaned['released_year'].value_counts()

# Display the frequencies
print("Artist(s) Frequency:")
print(artist_frequency)

print("\nReleased Year Frequency:")
print(year_frequency)
```

<h1><strong>3. Visualization:</strong></h1>

```python
# Set plot style
sns.set(style="whitegrid")

# Calculate artist(s) frequency
artist_frequency = df_cleaned['artist(s)_name'].value_counts()

# Select the top 10 artists
top10 = artist_frequency.head(10)

# Bar plot of top 10 artist(s) frequency with labels
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=top10.index, y=top10.values, palette="plasma")
plt.title('1. Top 10 Artists by Song Released in Spotify')
plt.xlabel('Artist(s)')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')

# Add labels to the bars
for i, value in enumerate(top10.values):
    ax.text(i, value + 0.1, str(value), ha='center', va='bottom')

plt.show()
```

The bar plot shows Taylor Swift's clear dominance in the Spotify landscape, with an impressive 34 songs released, securing her position as the artist with the most Spotify releases. The Weeknd comes in second with 22 songs. Bad Bunny and Sza are tied for third place, with each having released 19 songs on Spotify, adding to the platform's diverse musical offerings.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='artist_count', data=df, palette='viridis')
plt.title('2. Distribution of Songs by Number of Artists')
plt.xlabel('Number of Artists')
plt.ylabel('Count')
plt.show()
```

```python
# Convert 'released_year' to numeric using .loc
df_cleaned['released_year'] = pd.to_numeric(df_cleaned['released_year'], errors='coerce')

# Check for NaN values after the conversion
print("Number of NaN values in 'released_year' after conversion:", df_cleaned['released_year'].isna().sum())

# Drop rows with NaN values in 'released_year'
df_cleaned = df_cleaned.dropna(subset=['released_year'])

# Time series analysis - Number of songs released each year
plt.figure(figsize=(12, 6))

# Create the line plot
counts_per_year = df_cleaned.groupby('released_year').size().reset_index(name='count')
sns.lineplot(x='released_year', y='count', data=counts_per_year, marker='o', sort=False)

# Annotate each data point with its count
for line in range(0, counts_per_year.shape[0]):
     plt.text(counts_per_year['released_year'].values[line], counts_per_year['count'].values[line] + 10.0,
              str(counts_per_year['count'].values[line]), ha='center')

plt.title('3. Number of Songs Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.show()
```

The line plot shows an interesting pattern in Spotify song releases from the 1940s to the early 2000s, with a fairly narrow range of 1 to 23 songs during this time. However, a significant shift becomes visible beginning in 2020, as the number of releases increases dramatically. In 2020, the count increased dramatically to 67 songs, followed by a stunning increase to 119 in 2021. The momentum continued to build, reaching an impressive 402 releases in 2022 and, while slightly reduced, remaining significant at 175 songs in 2023. This increase in Spotify releases since 2020 suggests a dynamic and expanding landscape, which may reflect changes in the music industry, artist activity, or platform dynamics.


```python
# Box plot for streams
plt.figure(figsize=(10, 6))
sns.boxplot(x='streams', data=df_cleaned)
plt.title('4. Box Plot of Streams for Songs')
plt.xlabel('Streams')
plt.ylabel('Number of Streams')  # Add label for the y-axis
plt.show()
```

When interpreting the Box Plot of song streams, there is a noticeable positive skewness, which indicates a distribution with a mean greater than the median. Furthermore, the presence of outliers is notable, ranging from 1.5 to 3.0 and 3.5, indicating a subset of songs with significantly higher stream counts than the majority. This finding emphasizes the variability of streaming measures, potentially pointing to specific songs that have achieved exceptional popularity on the platform.

```python
# Calculate the top 10 Spotify playlists with the highest number of songs
top10_playlists = df_cleaned['in_spotify_playlists'].value_counts().head(10)

# Display the top 10 playlists
print("Top 10 Spotify Playlists with the Highest Number of Songs:\n", top10_playlists)

# Custom colors for the pie chart
custom_colors = ['lime', 'deeppink', 'yellow', 'aqua', 'coral']

# Pie chart for the top 10 playlists
plt.figure(figsize=(8, 8))
top10_playlists.plot.pie(autopct='%1.1f%%', startangle=90, colors=custom_colors)
plt.title('5. Top 10 Spotify Playlists with the Highest Number of Songs')
plt.ylabel('')
plt.show()
```

The pie chart shows that among the top ten playlists with the most songs, there is a remarkable balance, with each playlist containing exactly three songs. This even distribution demonstrates a balanced collection approach, with each playlist contributing equally to the representation of songs in the upper level. Such meticulous playlist management ensures a diverse and well-rounded listening experience across all top-ranked playlists.

```python
# Calculate the top 10 Apple playlists with the highest number of songs
top10_playlists = df_cleaned['in_apple_playlists'].value_counts().head(10)

# Display the top 10 playlists
print("Top 10 Apple Playlists with the Highest Number of Songs:\n", top10_playlists)

# Custom colors for the pie chart
custom_colors = ['chartreuse', 'pink', 'cyan', 'lightsalmon', 'magenta']

# Pie chart for the top 10 playlists
plt.figure(figsize=(8, 8))
top10_playlists.plot.pie(autopct='%1.1f%%', startangle=90, colors=custom_colors)
plt.title('6. Top 10 Apple Playlists with the Highest Number of Songs')
plt.ylabel('')
plt.show()
```

The earlier pie chart of the top ten Apple Playlists with the most songs shows a clear hierarchy. Playlist number 0 leads with an impressive 23 songs, followed by playlist number 8, which has 22 songs. As a result, playlists 10, 20, 4, and 7 share a tier of 20 songs each, demonstrating a notable consistency in song counts. Following this pattern, playlists 5 and 16 have 19 songs, while 11 and 3 have 17 songs each. This insightful breakdown provides a comprehensive view of the distribution of song counts across the top Apple Playlists, highlighting the organized collections' variation and varying popularity scales.


```python
# Set an aesthetically pleasing color palette
custom_palette = sns.color_palette("viridis", as_cmap=True)

# Set the style for visualizations
sns.set(style="whitegrid")

# Scatter plot for Streams vs. Spotify Charts Rank
plt.figure(figsize=(12, 8))
scatter_plot = sns.scatterplot(x='streams', y='in_spotify_charts', data=df, alpha=0.7, palette=custom_palette)

# Enhance plot aesthetics
scatter_plot.set_title('Scatter Plot: Streams vs. Spotify Charts Rank', fontsize=16, weight='bold')
scatter_plot.set_xlabel('Number of Streams', fontsize=12)
scatter_plot.set_ylabel('Spotify Charts Rank', fontsize=12)

# Customize grid lines
scatter_plot.grid(color='gray', linestyle='--', linewidth=0.5)

# Show a colorbar for better interpretation of data points
colorbar = plt.colorbar(scatter_plot.collections[0])
colorbar.set_label('Density of Points', fontsize=10)

# Add annotations or additional information if needed
# plt.annotate('Additional Information', xy=(x_coord, y_coord), xytext=(text_x, text_y),
#              arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10)

# Display the enhanced scatter plot
plt.show();
```

```python
# Dual axis plot for Spotify metrics
fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:cyan'
ax1.set_xlabel('Songs')
ax1.set_ylabel('Streams', color=color)
ax1.plot(df_cleaned['streams'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel('Apple Charts Rank', color=color)
ax2.plot(df_cleaned['in_apple_charts'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Comparison of Streams and Apple Charts Rank for Songs')
plt.show()
```

This dual-axis and scatter plot illustrates how the number of streams correlates with the song's position on the Spotify and Apple Music charts. Each data point represents a single song, and its position on the plot reflects the complex relationship between streams and chart rankings across both platforms.

Upon closer examination, an unexpected correlation appears demonstrating that higher chart rankings are consistently linked to increased streaming activity. The graph clearly shows that songs in the top 100 charts consistently receive significant streams, surpassing the impressive milestone of 3.5 billion. In contrary, songs outside the top 100 charts have lower stream counts, ranging from around 0.5 billion to just under 2.0 billion.

This fascinating observation sheds light on how important chart performance is in determining a song's streaming popularity on platforms such as Spotify. The data vividly illustrates that achieving higher chart positions significantly enhances a song's reach, attracting a much wider audience and developing greater success in the competitive landscape of music streaming.




```python
# Heatmap for correlation
plt.figure(figsize=(10, 8))
sns.heatmap(df_cleaned.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

Furthermore, a significant correlation of 0.71 exists between 'in_spotify_playlist' and 'in_apple_playlist.' The heatmap also shows a moderately positive correlation (0.55) between 'in_spotify_charts' and 'in_apple_charts.' These correlation coefficients emphasize the interrelationships of these variables, with especially strong associations between song streams and inclusion in Spotify and Apple Music playlists.
