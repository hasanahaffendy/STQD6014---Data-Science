<h2> <strong> About Dataset: </strong> </h2>
This dataset contains a comprehensive list of the most popular songs in 2023, as listed on Spotify. The dataset contains many more features than similar datasets. It provides information about each song's characteristics, popularity, and presence across various music platforms. The dataset contains the track name, artist(s) name, release date, Spotify playlists and charts, streaming statistics, and Apple Music.


<h3> <strong> Key Features: </strong> </h3>

- track_name: Name of the song

- artist(s)_name: Name of the artist(s) of the song

- artist_count: Number of artists contributing to the song

- released_year: Year when the song was released

- released_month: Month when the song was released

- released_day: Day of the month when the song was released

- in_spotify_playlists: Number of Spotify playlists the song is included in

- in_spotify_charts: Presence and rank of the song on Spotify charts

- streams: Total number of streams on Spotify

- in_apple_playlists: Number of Apple Music playlists the song is included in

- in_apple_charts: Presence and rank of the song on Apple Music charts

Link: https://www.kaggle.com/datasets/nelgiriyewithana/top-spotify-songs-2023

<h1><strong>Step 1: Data Cleaning</strong></h1>

```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
from google.colab import files
```

```python
# Example URL
spotify = 'spotify-2023.csv'

# Read data directly from the URL
df = pd.read_csv(spotify, encoding='latin1')
```

```python
# Print the DataFrame
print(df)
```

```python
# Check DataFrame Information
df.info()
```

```python
# Perform necessary cleaning steps
# Drop columns with NaN values
df_cleaned = df.dropna(axis=1)
df_cleaned
```

<h1><strong>2. Frequency Analysis:</strong></h1>

```python
# Frequency of songs by artist(s)
artist_frequency = df_cleaned['artist(s)_name'].value_counts()

# Frequency of songs by released year
year_frequency = df_cleaned['released_year'].value_counts()

# Display the frequencies
print("Artist(s) Frequency:")
print(artist_frequency)

print("\nReleased Year Frequency:")
print(year_frequency)
```

<h1><strong>3. Visualization:</strong></h1>

```python
# Set plot style
sns.set(style="whitegrid")

# Calculate artist(s) frequency
artist_frequency = df_cleaned['artist(s)_name'].value_counts()

# Select the top 10 artists
top10 = artist_frequency.head(10)

# Bar plot of top 10 artist(s) frequency with labels
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=top10.index, y=top10.values, palette="plasma")
plt.title('Top 10 Artists by Song Released in Spotify')
plt.xlabel('Artist(s)')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')

# Add labels to the bars
for i, value in enumerate(top10.values):
    ax.text(i, value + 0.1, str(value), ha='center', va='bottom')

plt.show()
```

```python
# Convert 'released_year' to numeric using .loc
df_cleaned['released_year'] = pd.to_numeric(df_cleaned['released_year'], errors='coerce')

# Check for NaN values after the conversion
print("Number of NaN values in 'released_year' after conversion:", df_cleaned['released_year'].isna().sum())

# Drop rows with NaN values in 'released_year'
df_cleaned = df_cleaned.dropna(subset=['released_year'])

# Time series analysis - Number of songs released each year
plt.figure(figsize=(12, 6))

# Create the line plot
counts_per_year = df_cleaned.groupby('released_year').size().reset_index(name='count')
sns.lineplot(x='released_year', y='count', data=counts_per_year, marker='o', sort=False)

# Annotate each data point with its count
for line in range(0, counts_per_year.shape[0]):
     plt.text(counts_per_year['released_year'].values[line], counts_per_year['count'].values[line] + 5.0,
              str(counts_per_year['count'].values[line]), ha='center')

plt.title('Number of Songs Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.show()
```

```python
# Box plot for streams
plt.figure(figsize=(10, 6))
sns.boxplot(x='streams', data=df_cleaned)
plt.title('Box Plot of Streams for Songs')
plt.xlabel('Streams')
plt.ylabel('Number of Streams')  # Add label for the y-axis
plt.show()
```

```python
# Calculate the top 10 playlists with the highest number of songs
top10_playlists = df_cleaned['in_spotify_playlists'].value_counts().head(10)

# Display the top 10 playlists
print("Top 10 Playlists with the Highest Number of Songs:\n", top10_playlists)

# Pie chart for the top 10 playlists
plt.figure(figsize=(8, 8))
top10_playlists.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Playlists with the Highest Number of Songs')
plt.ylabel('')
plt.show()
```

```python
# Dual axis plot for Spotify metrics
fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:purple'
ax1.set_xlabel('Songs')
ax1.set_ylabel('Streams', color=color)
ax1.plot(df_cleaned['streams'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:pink'
ax2.set_ylabel('Spotify Charts Rank', color=color)
ax2.plot(df_cleaned['in_spotify_charts'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Comparison of Spotify Streams and Spotify Charts Rank for Songs')
plt.show()
```

```python
# Heatmap for correlation
plt.figure(figsize=(10, 8))
sns.heatmap(df_cleaned.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
