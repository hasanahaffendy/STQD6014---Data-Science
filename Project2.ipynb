<h2> <strong> About Dataset: </strong> </h2>
This dataset contains a comprehensive list of the most popular songs in 2023, as listed on Spotify. The dataset contains many more features than similar datasets. It provides information about each song's characteristics, popularity, and presence across various music platforms. The dataset contains the track name, artist(s) name, release date, Spotify playlists and charts, streaming statistics, and Apple Music.


<h3> <strong> Key Features: </strong> </h3>

- track_name: Name of the song

- artist(s)_name: Name of the artist(s) of the song

- artist_count: Number of artists contributing to the song

- released_year: Year when the song was released

- released_month: Month when the song was released

- released_day: Day of the month when the song was released

- in_spotify_playlists: Number of Spotify playlists the song is included in

- in_spotify_charts: Presence and rank of the song on Spotify charts

- streams: Total number of streams on Spotify

- in_apple_playlists: Number of Apple Music playlists the song is included in

- in_apple_charts: Presence and rank of the song on Apple Music charts

```python
import pandas as pd

# Example URL
spotify = 'spotify-2023.csv'

# Read data directly from the URL
df = pd.read_csv(spotify, encoding='latin1')
```

```python
# Print the DataFrame
print(df)
```

```python
# Check DataFrame Information
df.info()
```

```python
# Perform necessary cleaning steps
# Drop columns with NaN values
df_cleaned = df.dropna(axis=1)
df_cleaned
```

<h1><strong>2. Visualizations:</strong></h1>

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='released_year', y='Count', data=df_cleaned)
plt.title('Bar Plot of Category Counts')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```
