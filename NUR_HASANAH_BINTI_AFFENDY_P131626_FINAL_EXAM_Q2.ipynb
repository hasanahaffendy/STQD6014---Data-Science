```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Ellipse, RegularPolygon

# Create a figure and axis
fig, ax = plt.subplots()

# Create a square
square_x = -1
square_y = 3
square_side_length = 4  # Area = side_length * side_length = 16
square = Rectangle((square_x, square_y), square_side_length, square_side_length, color='red')

# Add the square to the plot
ax.add_patch(square)

# Create an ellipse
ellipse_center_x = 0.5
ellipse_center_y = 1.5
ellipse_width = 3
ellipse_height = 5
ellipse = Ellipse((ellipse_center_x, ellipse_center_y), width=ellipse_width, height=ellipse_height, color='blue', alpha=0.5)

# Add the ellipse to the plot
ax.add_patch(ellipse)

# Create a hexagon
hexagon_vertices = [(3, 4), (5, 8), (7, 8), (9, 4), (7, 0), (5, 0)]
hexagon = RegularPolygon((5, 4), numVertices=6, radius=1, orientation=0, edgecolor='green', facecolor='none', linestyle='dotted')

# Add the hexagon to the plot
ax.add_patch(hexagon)

# Set limits for the axes
ax.set_xlim([-2, 10])  # Adjusted to include all shapes
ax.set_ylim([-1, 9])   # Adjusted to include all shapes

# Show the plot
plt.show()
```
