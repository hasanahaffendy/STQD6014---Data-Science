```python
def count_letter_in_file(file_path, target_letter):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            count = content.count(target_letter)
            return count
    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
        return 0

def main():
    target_letter = input("Enter the letter you want to count: ")
    file_paths = ["test1.txt", "test2.txt", "test3.txt"]

    total_count = 0
    for file_path in file_paths:
        count = count_letter_in_file(file_path, target_letter)
        print(f"Number of times '{target_letter}' appears in '{file_path}': {count}")
        total_count += count

    print(f"Total number of times '{target_letter}' appears in all files: {total_count}")

if __name__ == "__main__":
    main()
```

```python
# List of uploaded files
file_names = ["MobyDick.txt", "Pride_Prejudice.txt", "Frankenstein.txt"]

# Concatenate the contents of the files
concatenated_content = ""
for file_name in file_names:
    with open(file_name, 'r') as file:
        concatenated_content += file.read()

# Print the concatenated content
print(concatenated_content)
```

```python
#Counter counts the elements that has been passed in and returns a key, value pair with the element counts. (https://medium.com/analytics-vidhya/beginners-guide-from-collections-import-counter-d93bcecd190b)
from collections import Counter
import string
```

```python
# Function to clean the text by removing punctuation and converting to lowercase so that the text becomes more easily understandable for NLP models that are learning the text
# https://towardsdatascience.com/primer-to-cleaning-text-data-7e856d6e5791#:~:text=Text%20cleaning%20here%20refers%20to,reducing%20noise%20in%20text%20data.
def clean_text(text):
    translator = str.maketrans("", "", string.punctuation)
    cleaned_text = text.translate(translator)
    return cleaned_text.lower()
```

```python
# Function to get the top ten keywords and their counts and percentages
def get_top_keywords(text):
    # List of common words to exclude
    exclude_common_words = set(["a", "the", "and", "to", "from", "where", "what", "was", "when", "I", "you", "we", "us", "me", "for"])

    # Split the cleaned text into words and filter out common words
    words = clean_text(text).split()
    filtered_words = [word for word in words if word not in exclude_common_words]

    # Count the occurrences of each remaining word
    word_counts = Counter(filtered_words)

    # Get the top ten keywords and calculate their percentages
    top_keywords = word_counts.most_common(10)
    total_words = len(filtered_words)
    percentages = [(word, count, (count / total_words) * 100) for word, count in top_keywords]

    return percentages
```

```python
# Function to count occurrences of a specific keyword
def count_specific_keyword(text, keyword):
    cleaned_text = clean_text(text)
    word_counts = Counter(cleaned_text.split())
    return word_counts[keyword.lower()]
```

```python
# Get and print the top ten keywords
top_keywords = get_top_keywords(concatenated_content)
print("Top Ten Keywords:")
for word, count, percentage in top_keywords:
    print(f"{word}: Count - {count}, Percentage - {percentage:.2f}%")
```
