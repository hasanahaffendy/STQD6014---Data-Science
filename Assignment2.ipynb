```python
class BankAccount:
    def __init__(self, username, full_name, ic_number, city, telephone_number, age, parent_username=None):
        self.username = username
        self.full_name = full_name
        self.ic_number = ic_number
        self.city = city
        self.telephone_number = telephone_number
        self.balance = 20  # Initial balance with a minimum requirement of RM20
        self.age = age
        self.parent_username = parent_username  # Only for child accounts
        self.account_type = self.determine_account_type()

    def determine_account_type(self):
        if self.age > 18:
            return "Parent"
        else:
            return "Child"

    def deposit(self):
        amount = float(input(f"\nEnter the deposit amount for {self.username}: RM "))
        if amount > 0:
            self.balance += amount
            print(f"\nDeposited RM{amount}. New balance: RM{self.balance} ")
        else:
            print("\nInvalid deposit amount.")

    def withdraw(self):
        amount = float(input(f"\nEnter the withdrawal amount for {self.username}: RM "))
        if amount > 0:
            if self.balance - amount >= 20:  # Enforcing minimum balance requirement
                self.balance -= amount
                print(f"\nWithdrew RM{amount}. New balance: RM{self.balance} ")
            else:
                print("\nInsufficient funds. Minimum balance requirement not met.")
        else:
            print("\nInvalid withdrawal amount.")

    def check_balance(self):
        print(f"\nCurrent balance for {self.username}: RM{self.balance} ")

    def close_account(self):
        print(f"\nAccount for {self.username} closed. Remaining balance: RM{self.balance}")


# Function to create a bank account based on user input
def create_bank_account():
    username = input("\nEnter username: ")
    full_name = input("\nEnter full name: ")
    ic_number = input("\nEnter IC number: ")
    city = input("\nEnter city: ")
    telephone_number = input("\nEnter telephone number: ")
    age = int(input("\nEnter your age: "))

    if age <= 18:
        parent_username = input("\nEnter parent's username: ")
        return BankAccount(username, full_name, ic_number, city, telephone_number, age, parent_username)
    else:
        return BankAccount(username, full_name, ic_number, city, telephone_number, age)


# Testing scenarios
while True:
    account = create_bank_account()

    while True:
        action = input("\nDo you want to deposit (D), withdraw (W), check balance (B), or close account (C)? ").upper()
        if action == "D":
            account.deposit()
        elif action == "W":
            account.withdraw()
        elif action == "B":
            account.check_balance()
        elif action == "C":
            account.close_account()
            break
        else:
            print("\nInvalid action. Please enter D, W, B, or C.")

    continue_program = input("\nDo you want to create another account? (y/n): ").lower()
    if continue_program != 'y':
        break
```
